# Create dataframe using a dictionary

```python
dict1 = {'column A':[1,2,3,4,5],'ColumnB':[10,20,30,40,50],'ColumnC':[20,50,70,90,100]}
dict1
```

```python
df = pd.DataFrame(dict1)
df
```

```python
# Add columns using a list
df['ColumnD'] = [100,200,300,400,500]
df
```

```python
#Add column using array
import numpy as np
```

```python
np.array([50,40,30,70,100])
```

```python
df['ColumnE'] = np.array([50,40,30,70,100])
df
```

```python
#Add column using series
pd.Series([20.5,34,'python',3+4j,100])
```

```python
df['ColumnF'] = pd.Series([20.5,34,'python',3+4j,100])
df
```

```python
# Add a single value
df['ColumnG'] = 1000
df
```

# Create a dataframe using array

```python
# create a 5x4 array
array1 = np.random.randint(1,10,size=(5,6))
array1
```

```python
# create a dataframe of this array
pd.DataFrame(array1)
```

```python
# using columns and index parameter of dataframe give the column and index names

pd.DataFrame(array1,columns=['A','B','C','D','E','F'],index=['L','M','N','O','P'])
```

# Create a dataframe using list

```python
list1 = ['python',10,18.5,125,'data']
list1
```

```python
pd.DataFrame(list1,columns=['A'],index=['L','M','N','O','P'])
```
