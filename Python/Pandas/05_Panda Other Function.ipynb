# 1.Value_counts

```python
>> It returns a series of unique value counts

syntax>> df[categorical_column_name].value_counts()

Note - Dont use value_counts() function on any numerical column. Use this function only on
       categorical column.
```

```python
# Find value counts of STATE column
df['STATE'].value_counts()
```

# 7.Rename

```python
>> This function is used to rename the columns or rows in a dataframe

Syntax:
df.rename({oldname1:newname1,oldname2:newname2},axis=0)  >> To rename 1 or 2 row names
df.rename({oldname1:newname1,oldname2:newname2},axis=1)  >> To rename 1 or 2 column names

df.columns = column_names_list   >> To rename all column names
df.index   = row_names_list      >> To rename all row names
```

```python
df = pd.read_csv('../assets/50_Startups.csv')
df
```
