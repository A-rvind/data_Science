# Types of joins in DataFrame

# 1. Merge

```python
>> It is used to join 2 tables
>> To join 2 tables there must be a common column in 2 tables

Syntax >> pd.merge(left_df,right_df,left_on,right_on,how)
```

```python
# Read the emp_2.csv
left_df = pd.read_csv('../assets/emp_2.csv')
left_df
```

```python
# Read the dept.csv
right_df = pd.read_csv('../assets/dept.csv')
right_df
```

```python
# inner join
pd.merge(left_df,right_df,left_on='did',right_on='dept_id')  # default - inner join
```

```python
# left join
pd.merge(left_df,right_df,left_on='did',right_on='dept_id',how='left')
```

```python
# right join
pd.merge(left_df,right_df,left_on='did',right_on='dept_id',how='right')
```

```python
# outer join
pd.merge(left_df,right_df,left_on='did',right_on='dept_id',how='outer')
```

# 2.Concat

```python
>> It is used to find union of 2 dataframes

syntax >> pd.concat([df1,df2,df3 ...],axis=0,join)
```

```python
# Read the 50_Startups.csv
df = pd.read_csv('../assets/50_Startups.csv')
df
```

```python
# Create 2 dataframes to concat
A = df[0:5]  # first 5 rows
A
```

```python
B = df[5:10]  # Next 5 rows
B
```

```python
C = df[10:13]
C
```

```python
syntax >> pd.concat([df1,df2,df3 ...],axis=0,join)
```

```python
# Apply concat function
pd.concat([C,A,B])
```

```python
pd.concat([C,A,B],axis=1) #all column
```
