# 3. Sorting

```python
1. sort_values function

>> This function is used to sort the dataframe based on a column values
    ascending(lower-higher)(a-z)
    descending(higher-lower)(z-a)

syntax>> df.sort_values(column_name,ascending=True)  # Sort the dataframe in ascending order
         df.sort_values(column_list,ascending=True)
    
         df.sort_values(column_name,ascending=False) # Sort the dataframe in descending order

ascending = True  >> ascending order (default)
ascending = False >> descending order
```

```python
# Read the 50_Startups.csv
df = pd.read_csv('../assets/50_Startups.csv')
df
```

```python
# Sort the df based on RND column in ascending order
df.sort_values('RND',inplace=True)   # ascending = True >> Default
```

```python
df
```

```python
# Reset the index
df.index = range(0,50)
```

```python
df
```

```python
# Sort the df based on MKT column in descending order
df.sort_values('MKT',ascending=False,inplace=True)
df
```

```python
df.sort_values('STATE',inplace=True)
df
```

```python
# Sort the df based on STATE,PROFIT column in ascending order
df.sort_values(['STATE','PROFIT'],inplace=True)
df
```

```python
# Sort the df based on STATE,PROFIT column in descending order
df.sort_values(['STATE','PROFIT'],ascending=False,inplace=True)
df
```
