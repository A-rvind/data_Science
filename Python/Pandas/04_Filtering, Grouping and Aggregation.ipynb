```python
# Read the credit.csv
Drop Unnamed:0 and ID columns and create a new DF
1.Find out total number of people who have 2 cards
2.Find out total number of people who have 2 cards and are Not Married
3.Find out Genderwise average Income
4.Find out Ethinicity wise count of candidates
5.Find out total number of married and unmmaried candidates
```

```python
df2 = pd.read_csv('../assets/Credit.csv')
df2
```

```python
df2.drop('Unnamed: 0',axis=1,inplace=True)
df2
```

```python
# 1.Find out total number of people who have 2 cards
df2['Cards'][df2['Cards']==2].count()
```

```python
# 2.Find out total number of people who have 2 cards and are Not Married
df2['Cards'][(df2['Cards']==2) & (df2['Married']=='No')].count()
```

```python
#3.Find out Genderwise average Income
df2.groupby("Gender")["Income"].mean()
```

```python
#4.Find out Ethinicity wise count of candidates
df2.groupby("Ethnicity")["Income"].count()
```

```python
#5.Find out total number of married and unmmaried candidates
df2.groupby("Married")["Income"].count()
```
